```python
# import huggingface
import transformers
import torch
import numpy as np
import pandas as pd
import os
import json
import time
import random
import gc
import matplotlib.pyplot as plt
from torch.xpu import device
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM
from datasets import load_dataset
from huggingface_hub import notebook_login
from benchmarking_utils import evaluate_gsm8k
import re
```

```python
# Let us initialize the Qwen 500M model for benchmarking as it was what we mentioned on the report.

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-0.5B")
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen2.5-0.5B")
model.to("cuda")

# Define the pad token to be the EOS token
tokenizer.pad_token = tokenizer.eos_token
```

```python
# Id the math dataset that we planned on using for benchmarking
# The math dataset mentioned in the proposal is big-math, so we will be using that

# Load the dataset in a streaming fashion from huggingfac

gsm8k = load_dataset("gsm8k", "main", split="test")
```

```python
res = evaluate_gsm8k(model, tokenizer, gsm8k)
print(res)
```

For base Qwen-0.5B model, we got an accuracy of 31.74% on the GSM8k test dataset using 3-shot prompting. This is comparable to the results for Qwen-1.8B model on the same dataset, as seen here https://github.com/QwenLM/Qwen . Qwen 1.8B was able to an accuracy of 32.3 % with 8-shot prompting.

While the base version of the model struggles to perform on this dataset, from the same link we can see that there are models capable of performing well on this dataset ( For instance, the 7B version of Qwen was able to achieve an accuracy of 51.7% with 8-shot prompting. This is encouraging, and it shows us that fine-tuning methods, such as GRPO and PPO, can help us improve the performance of the base version of the model.
